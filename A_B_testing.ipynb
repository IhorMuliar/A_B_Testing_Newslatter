### Objective

Our objective here is to check which newsletter brings higher traffic on the website.

### 1. State the Hypothesis

<b>1.1 Null Hypothesis</b>

H is - there is no difference in the conversion rate in customers receiving newslatter A and B (A > B)

<b>1.2 Alternative Hypothesis</b>

H is - the conversion rate of newsletter B is higher than those who receive newsletter A (B > A)

### 2. Importing Libaries

```python
import pandas as pd
import numpy as np
import scipy.stats 

import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="white", palette="muted", color_codes=True)
sns.set_style('ticks')
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

### 3. EDA

```python
data = pd.read_csv("data.csv")
```

```python
sns.distplot(data.Conversion_B, color='y')
plt.tight_layout()
```

### 4. T-test

```python
t_stat, p_val= scipy.stats.ttest_ind(data.Conversion_B,data.Conversion_A)
print('t-score: {}, p-value: {}'.format(t_stat, p_val))
```

Critical Value = 2.0452

T-score = 3.787

P-value = 0.00036

T-score > Critical Value

P-value < 0.05

<b>Thus, P-value is less than 0.05 so we can reject the null hypothesis. This means that in our A/B testing, newsletter B is performing better than A.</b>
